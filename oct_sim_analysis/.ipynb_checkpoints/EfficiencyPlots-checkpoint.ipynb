```python
import os

import numpy as np
import uproot
import matplotlib.pyplot as plt
import modplot
```

```python
plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 120
plt.rcParams['font.family'] = 'serif'
plt.rcParams['text.usetex'] = False
```

## Loading the Data

```python
data_path = '/Users/anthonybadea/Documents/ATLAS/data/simResults_051920_102540'

nEvents = [10**4]
bkgRates = [i*(10**j) for i in range(1,10) for j in range(2,5)] 
bkgRates.sort()

real_x_muon = []
real_y_muon = []
trig_x = []
trig_y = []

# populate the data dictionary of dictionaries 
# with key: data --> (nEvent,bkgRate): single file data dictionary
# single file data dict key = [real_x_muon,real_y_muon,trig_x,trig_y] : data
data_dict = {}

# loop over the data files
for nE in nEvents:
    for bR in bkgRates:
        file_path = "job_{0}_{1}/outFile_nEvents{0}_bkgRate{1}.root".format(nE,bR)
        data_file = uproot.open(os.path.join(data_path,file_path))
        real_x_muon = data_file['gingko']['real_x_muon'].array()
        real_y_muon = data_file['gingko']['real_y_muon'].array()
        trig_x = data_file['gingko']['trig_x'].array()
        trig_y = data_file['gingko']['trig_y'].array()
        data_dict[(nE,bR)] = {'real_x_muon': real_x_muon,
                             'real_y_muon': real_y_muon,
                             'trig_x': trig_x,
                             'trig_y': trig_y}
```

```python
# compute the distributions to determine the bin-by-bin corrections
bins = np.linspace(0, 3500, 350)
density = False

#plt.hist(nature['data_thrusts'], bins=bins, histtype='step', color='black', density=density)
real_x_muon_hist = plt.hist(data_dict[(10**4,200)]['real_x_muon'], bins=bins, histtype='step', color='black', density=density, label='Real Muon X')[0]
real_y_muon_hist = plt.hist(data_dict[(10**4,200)]['real_y_muon'], bins=bins, histtype='step', color='blue', density=density, label='Real Muon Y')[0]
#trig_x_hist = plt.hist(trig_x, bins=bins, histtype='step', color='red', density=density)[0]
#trig_y_hist = plt.hist(trig_y, bins=bins, histtype='step', color='green', density=density)[0]

#plt.yscale('log')
plt.legend(loc="upper right")
plt.show()
```

```python
# Calculate the number of triggered muons 
xCut = 5
yCut = 5
for nE in nEvents:
    for bR in bkgRates:
        real_x_muon = data_dict[(nE,bR)]['real_x_muon']
        real_y_muon = data_dict[(nE,bR)]['real_y_muon']
        trig_x = data_dict[(nE,bR)]['trig_x']
        trig_y = data_dict[(nE,bR)]['trig_y']
        TObs = 0
        for i in range(len(real_x_muon)):
            realMuonX = real_x_muon[i]
            realMuonY = real_y_muon[i]
            for j in range(len(trig_x[i])):
                trigX = trig_x[i][j]
                trigY = trig_y[i][j]
                if (abs(realMuonX - trigX) < xCut) and (abs(realMuonY - trigY) < yCut):
                    TObs+=1
                    break     
        data_dict[(nE,bR)]['TObs'] = TObs
```

```python
# Compute the trigger efficiencies
Ed = 1.0
P = (1-(1-Ed)**2)**4
ETrigs = []
ETrigs_errs = []
for nE in nEvents:
    for bR in bkgRates:
        TObs = data_dict[(nE,bR)]['TObs']
        NExp = P*nE
        ETrigs.append( TObs / NExp)
        ETrigs_errs.append((1/NExp)*(TObs * (1-TObs/NExp))**0.5)
```

```python
# a dictionary to hold information about the observables
obs = {}
obs.setdefault('EffVsBkgRate',{}).update({
    'xlim': (0.5*10**2,10**5), 'ylim': (0.35,0.9), 'ylim_ratio': (0.7,1.3),
    'xlabel': 'Background Rate (Hz)', 'ylabel': 'Trigger Efficiency', 'ylabel_ratio':"TEST",
    'legend_loc': 'lower left', 'legend_ncol': 1
})

effvsBkgRateStyle = {'ls': '-', 'marker': 'o', 'ms': 3.5, 'color': 'black', 'zorder': 1}
legend_opts = {'handlelength': 2.0, 'loc': 'lower left', 'frameon': False, 'numpoints': 1}
```

```python
for i,(obkey,ob) in enumerate(obs.items()):
    # get the styled axes on which to plot
    fig, [ax0, ax1] = modplot.axes(**ob)
    
    ax0.errorbar(bkgRates, ETrigs, yerr=ETrigs_errs, **effvsBkgRateStyle, label='Trigger Effiency')
    ax1.plot(ob['xlim'], [1, 1], '--', color='green', lw=0.75)
    
    # additional axis settings
    ax0.set_xscale("log")
    ax0.grid()
    ax1.set_xscale("log")
    #ax1.grid()
    fig.suptitle("Detector Efficiency = 1",fontsize=10)
    
    # legend style and ordering
    loc, ncol = ob.get('legend_loc', 'upper right'), ob.get('legend_ncol', 1)
    order = [0]
    
    # plot the legend
    ax0.legend(**legend_opts)
    
    fig.savefig('{}.pdf'.format(obkey), bbox_inches='tight')
    plt.show()
```
